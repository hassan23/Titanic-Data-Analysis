## Discription

The sinking of the RMS Titanic is one of the most infamous shipwrecks in history.  On April 15, 1912, during her maiden voyage, the Titanic sank after colliding with an iceberg, killing 1502 out of 2224 passengers and crew. This sensational tragedy shocked the international community and led to better safety regulations for ships.

One of the reasons that the shipwreck led to such loss of life was that there were not enough lifeboats for the passengers and crew. Although there was some element of luck involved in surviving the sinking, some groups of people were more likely to survive than others, such as women, children, and the upper-class.

To performe Data analysis on the give data which is a sample of the orignal data.

## Data extraction and cleaning

```python
# Rendering plots inline
%matplotlib inline

# Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# uploading the data as data frames
titanic_data = pd.read_csv('titanic-data.csv')
```

```python
titanic_data.head()
```

```python
titanic_data.tail()
```

### Removing duplicates
we first remove the duplicate entries(if any) from the data set.

```python
print 'There are total {} rows in the data set'.format(len(titanic_data))
# removing the duplicate entries
titanic_clean_data = titanic_data.drop_duplicates()

print 'There are total {} rows in the data set'.format(len(titanic_clean_data))
```

We see that the row count is still the same, that means there were not any duplicates in the data.

There are many entries which hava age missing. So we are going to see how many rows are there with NaN age and how it will effect our data analysis.

```python
missing_age = titanic_clean_data['Age'].isnull()
print '{} passangers whose age is missing.'.format(missing_age.sum())
print '{} male passangers whose age is missing.'.format((titanic_clean_data[missing_age]['Sex'] == 'male').sum())
print '{} female passangers whose age is missing.'.format((titanic_clean_data[missing_age]['Sex'] == 'female').sum())
```

So missing ages are a lot, it will effect out data analysis, especially for males.

## Analyse the data by grouping into age group

In this part we are going to group all the passangers into three groups

- child (less than 18)
- adult (between 18 to 60)
- old (above 60)

we did this kind of analysis above but this time we are going to make distinction about who is child and who is adult and who is old.

```python
def group_by_age(age):
    if age < 18:
        return 'child'
    elif age > 60: 
        return 'old'
    else:
        return 'adult'

titanic_more_clean_data['AgeGroup'] = titanic_more_clean_data['Age'].apply(group_by_age)   
```

```python
titanic_more_clean_data.head()
```

```python
sns.factorplot('AgeGroup', 'Survived',data = titanic_more_clean_data, ci=None, size=6, kind="bar")
```

Here we can see that children are definetly the one who have the highest survival rate followed by adults and the last old passangers

##### lets see how many childrens are with out there parents and how it will effect their survival

```python
children_without_parent = titanic_more_clean_data[(titanic_more_clean_data['AgeGroup'] == 'child') & 
                                                  (titanic_more_clean_data['Parch'] == 0)]
children_with_parent = titanic_more_clean_data[(titanic_more_clean_data['AgeGroup'] == 'child') & 
                                                  (titanic_more_clean_data['Parch'] > 0)]
print "Survival percentage of childrens with their parents is {}".format(children_with_parent.mean()['Survived']*100)
print "Survival percentage of childrens without their parents is {}".format(children_without_parent.mean()['Survived']*100)
```

```python
sns.factorplot('Survived',data = children_without_parent, ci=None, size=6, kind="count")
sns.plt.title('Childrens with out parents')
sns.factorplot('Survived',data = children_with_parent, ci=None, size=6, kind="count")
sns.plt.title('children with parents')
```

```python
fig, ax = plt.subplots(1,2)
sns.violinplot(x="Survived", y="Age", data=children_without_parent,ax=ax[0])
ax[0].set_title('Survival/Deal of Children without parent')
sns.violinplot(x="Survived", y="Age", data=children_with_parent,ax=ax[1])
ax[1].set_title('Survival/Deal of Children with parent')
```

From this we can easily say that the childrens who are **without** their parents have not survived in much percentage as they survived with parents. It might be because they were **with** their caretakers/nanies who at the time of calamity thought for their own life rather than the child. Or they must be travelling alone.

** Now we are going to see if old people has more survival rate if they were with their siblings**

```python
old_with_sibling = titanic_more_clean_data[(titanic_more_clean_data['AgeGroup'] == 'old') & 
                         (titanic_more_clean_data['SibSp'] > 0)]
old_not_with_sibling = titanic_more_clean_data[(titanic_more_clean_data['AgeGroup'] == 'old') & 
                         (titanic_more_clean_data['SibSp'] == 0)]
print "Survival percentage of old passangers with the siblings is {}".format(old_with_sibling.mean()['Survived']*100)
print "Survival percentage of old passangers without siblings is {}".format(old_not_with_sibling.mean()['Survived']*100)
```

```python
sns.factorplot('Survived',data = old_with_sibling, ci=None, size=6, kind="count")
sns.plt.title('survived/dead Old with Siblings')
sns.factorplot('Survived',data = old_not_with_sibling, ci=None, size=6, kind="count")
sns.plt.title('survived/dead Old without Siblings')
```

Here we can see a relevant difference between survival of old passangers who were and were not with their siblings, the surivial rate for old passangers with their sibling is significantly higher than the old passangers who were not with their siblings.

**Now Lets see the adult passangers and their surivival rate with or without their siblings**

```python
adult_with_sibling = titanic_more_clean_data[(titanic_more_clean_data['AgeGroup'] == 'adult') & 
                         (titanic_more_clean_data['SibSp'] > 0)]
adult_not_with_sibling = titanic_more_clean_data[(titanic_more_clean_data['AgeGroup'] == 'adult') & 
                         (titanic_more_clean_data['SibSp'] == 0)]

print "Survival percentage of old passangers with the siblings is {}".format(adult_with_sibling.mean()['Survived']*100)
print "Survival percentage of old passangers with the siblings is {}".format(adult_not_with_sibling.mean()['Survived']*100)
```

```python
sns.factorplot('Survived',data = adult_with_sibling, ci=None, size=6, kind="count")
sns.plt.title('survived/dead Adult with Siblings')
sns.factorplot('Survived',data = adult_not_with_sibling, ci=None, size=6, kind="count")
sns.plt.title('survived/dead Adult without Siblings')
```

Here we can see a relevant difference between survival of adult passangers who were and were not with their siblings, the surivial rate for adult passangers with their sibling is significantly higher than the adult passangers who were not with their siblings.
